```python
import matplotlib.pyplot as plt
```

```python
img = plt.imread('IMG_1.jpg')
plt.clf()
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
img = plt.imread('IMG_1.jpg')
print(img.shape)
plt.clf()
plt.imshow(img[:,:,0],cmap='Reds',vmin=0,vmax=255)
plt.colorbar()
plt.show()
```
